```python
import pandas as pd
from textblob import TextBlob
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
# Load the CSV file
df = pd.read_csv('/content/Elon_musk.csv')
```

```python
# Create a new column 'Sentiment' to store sentiment analysis results
df['Sentiment'] = df['Text'].apply(lambda x: TextBlob(str(x)).sentiment.polarity)
```

```python
# Classify sentiments into Positive, Negative, or Neutral
df['Sentiment_Label'] = df['Sentiment'].apply(lambda x: 'Positive' if x > 0 else 'Negative' if x < 0 else 'Neutral')
```

```python
# Display the DataFrame with sentiment analysis results
print(df[['Text', 'Sentiment_Label']])
```

```python
# Plotting the sentiment distribution
plt.figure(figsize=(8, 6))
df['Sentiment_Label'].value_counts().plot(kind='bar', color=['green', 'red', 'gray'])
plt.title('Sentiment Distribution')
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.show()
```

```python
# Positive word cloud
positive_tweets = df[df['Sentiment_Label'] == 'Positive']['Text'].str.cat(sep=' ')
wordcloud_positive = WordCloud(width=800, height=400, background_color='white').generate(positive_tweets)

plt.figure(figsize=(10, 6))
plt.imshow(wordcloud_positive, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud - Positive Tweets')
plt.show()
```

```python
# Negative word cloud
negative_tweets = df[df['Sentiment_Label'] == 'Negative']['Text'].str.cat(sep=' ')
wordcloud_negative = WordCloud(width=800, height=400, background_color='white').generate(negative_tweets)

plt.figure(figsize=(10, 6))
plt.imshow(wordcloud_negative, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud - Negative Tweets')
plt.show()
```

```python
# Save the DataFrame with sentiment analysis results to a new CSV file
output_file_path = 'Elon_musk_sentiment_analysis.csv'
df.to_csv(output_file_path, index=False)

print(f"Sentiment analysis completed and results saved to {output_file_path}.")
```
